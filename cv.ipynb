```python
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df.drop("Final_Result", axis=1)
y = df["Final_Result"]

X.head(), y.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    model,
    X_scaled,
    y,
    cv=kf,
    scoring="accuracy"
)

scores
```

```python
print("Accuracy for each fold:", scores)
print("Mean Accuracy:", scores.mean())
print("Standard Deviation:", scores.std())
```

```python
for k in [3, 5, 7, 10]:
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_scaled, y, cv=kf)
    print(f"K = {k} â†’ Mean Accuracy = {scores.mean():.4f}")
```

```python
## Conclusion

- K-Fold Cross Validation improves model reliability.
- Logistic Regression performs consistently on the dataset.
- Mean accuracy indicates overall performance.
- Standard deviation shows model stability.
- Cross Validation helps prevent overfitting.
```
